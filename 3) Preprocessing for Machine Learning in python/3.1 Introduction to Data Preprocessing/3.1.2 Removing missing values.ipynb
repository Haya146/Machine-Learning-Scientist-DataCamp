## Dataset

```python
import pandas as pd
import numpy as np

# Define the dataset
data = {
    'A': [1.0, 4.0, 7.0, np.nan, 5.0],
    'B': [np.nan, 7.0, np.nan, 7.0, 9.0],
    'C': [2.0, 3.0, np.nan, np.nan, 7.0]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df.isna().sum()
```

## Different ways for removing missing data

```python
df.dropna()
```

```python
df.drop([1,2,3])
```

```python
df.drop("A", axis=1)
```

```python
df
```

```python
# Only remove the missing values in column B
df.dropna(subset=["B"])
```

```python
df
```

```python
# drops rows that have fewer than 2 non-NaN (non-missing) values.
df.dropna(thresh=2)
```
