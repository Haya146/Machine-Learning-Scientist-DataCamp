## DataSet

```python
import pandas as pd
import numpy as np

# Define the dataset
data = {
    'A': [1, 2, 3],
    'B': ['string', 'string2', 'string3'],
    'C': ['1.0','2.0', '3.0']
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

- int64 : integer

- datetime64 : dates and times
