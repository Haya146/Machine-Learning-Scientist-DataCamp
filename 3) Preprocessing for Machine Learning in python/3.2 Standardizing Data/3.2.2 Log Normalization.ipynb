# What is log normalization?       
                 
- Useful for features with high variance      
- Applies logarithm transformation     
- Natural log using the constant e (â‰ˆ 2.718)       
- Captures relative changes, the magnitude of change, and keeps everything positive   

```python
import pandas as pd
import numpy as np

# Define the dataset
data = {
    'col1': [1.00, 1.20, 0.75, 1.60],
    'col2': [3.0, 45.5, 28.0,100.0]
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df.var()
```

```python
df['log_2'] = np.log(df['col2'])
df
```
